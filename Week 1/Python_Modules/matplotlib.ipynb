```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Line Plot

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y,color='blue',linestyle='--',marker='o',markersize=20,markerfacecolor='pink',markeredgecolor='red')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#### Multiple Line Plot

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Plot multiple lines
plt.plot(x, y1, label="y = x^2", color="blue", linestyle="--", marker="o")
plt.plot(x, y2, label="y = x", color="red", linestyle="-.", marker="s")

# Add labels, legend, and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()

# Show the plot
plt.show()
```

#### Bar Plot

```python
x = ['A','B','C','D']
y = [40,10,50,20]

plt.bar(x,y,color='skyblue')
plt.xlabel("Categories")
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
plt.barh(x,y,color='pink')
plt.ylabel("Categories")
plt.xlabel('Values')
plt.title('Horizontal Bar Plot')
plt.show()
```

#### Histogram

```python
data = np.random.randn(1000) # data contains thousand values from the standard normal distribution i.e. N(0,1)
plt.hist(data,bins=50,color='skyblue',edgecolor='black')
plt.title("Histogram")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show() 
```

#### Scatter Plot

```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100) # Color values
plt.scatter(x,y,c=colors,cmap='viridis')
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar()  # Adding the color bar
plt.show()
```

#### Pie Chart

```python
# Data
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 30, 20, 10]
colors = ["gold", "lightblue", "lightgreen", "pink"]
explode = (0.1, 0, 0, 0)  # Highlight the first slice

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True)

# Add title
plt.title("Programming Language Popularity")

# Show the plot
plt.show()
```

#### Box Plot

```python
data = np.array([7, 15, 13, 10, 22, 30, 5, 8, 15])
plt.boxplot(data)
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```

Here the end points of the wishkers(i.e. the T shaped lines coming out of the box) represents the minimum and maximum values excluding the outliers . The circles at the top and the bottom represents the outliers. the orange line represent the middle value (i.e. 50th quantile)

#### Area Plot

```python
days = [1,2,3,4]
eating = [3,4,2,4]
sleeping = [8,7,8,9]

plt.stackplot(days,eating,sleeping,labels=['Eating','Sleeping'])
plt.title("Stack Plot")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.legend(loc='upper left')
plt.show()
```
