```python
import pandas as pd
import numpy as np
```

```python
data = {'Name': ['Aman','Harsh','Shivam','Ayush','Ankit','Raj','Manish','Rahul','Apoorv','Aakash'],
        'Subject':['Physics','Chemistry','Maths','Chemistry','Maths','Physics','Chemistry','Chemistry','Maths','Maths'],
        'Score':np.random.randint(50,101,10),
        'Grade':['']*10}
df = pd.DataFrame(data)
df
```

```python
df.loc[df['Score']>=90,'Grade'] = 'A'
df.loc[(df['Score']>=80) & (df['Score']<90),'Grade'] = 'B'
df.loc[(df['Score']>=70) & (df['Score']<80),'Grade'] = 'C'
df.loc[(df['Score']>=60) & (df['Score']<70),'Grade'] = 'D'
df.loc[(df['Score']<60),'Grade'] = 'F'
df
```

```python
df_sorted = df.sort_values(by='Score',ascending=False)
df_sorted
```

```python
subject_avg = df.groupby(by='Subject')['Score'].mean()
subject_avg
```

```python
def pandas_filter_pass(dataframe):
    df = dataframe.loc[(dataframe['Grade']=='A') | (dataframe['Grade']=='B')]
    return df
```

```python
filtered_dataframe = pandas_filter_pass(df)
filtered_dataframe
```
